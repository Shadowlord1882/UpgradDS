```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import statsmodels
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import MinMaxScaler

# Read training data
train = pd.read_csv("https://media-doselect.s3.amazonaws.com/generic/QNdMORzykKkG4L3WQ17Xx53o3/insurance_training.csv")

# Read test data
test = pd.read_csv("https://media-doselect.s3.amazonaws.com/generic/Mq5p8YpQoz3KbWJRE4Mey1Yoq/insurance_test.csv")
train=train.drop('Unnamed: 0',1)
scaler = MinMaxScaler()
train_col=['charges','age', 'bmi']
new_train=train

new_train[train_col] = scaler.fit_transform(new_train[train_col])
# print(new_train.head())
#new_train=new_train.drop('BMI_group',1)
#new_train=new_train.drop('region',1)
#new_train=new_train.drop('sex',1)
#new_train=new_train.drop('bmi',1)
#new_train=new_train.drop('age',1)
#new_train=new_train.drop('children',1)
# new_train=new_train.drop('charges',1)
X_train = new_train
y_train = new_train.pop('smoker')


X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()
print(lr.summary())


vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
